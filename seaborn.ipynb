# Seaborn

## Seaborn Line Plot

```python
import seaborn as sns
from matplotlib import pyplot as plt

fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
sns.lineplot(x="timepoint", y="signal", data=fmri)
plt.show()
```

```python
# Grouping data with hue

sns.lineplot(x="timepoint",y="signal",data=fmri,hue="event")
plt.show()
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri,hue="event",style="event")
plt.show()
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri,hue="event",style="event",markers=True)
plt.show()
```

## Seaborn bar plot

```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt

sns.set(style="whitegrid")
iris=pd.read_csv('../input/iris-dataset-new/iris.csv')
sns.barplot(x="Sales",y="Profit",data=iris)
plt.show()
```

```python
sns.barplot(x="Sales",y="Profit",hue="Price",data=iris)
plt.show()
```

```python
sns.barplot(x="Sales",y="Profit",data=iris,palette='Blues_d')
plt.show()
```

```python
sns.barplot(x="Sales",y="Profit",data=iris,palette='rocket')
plt.show()
```

```python
sns.barplot(x="Sales",y="Profit",data=iris,palette='vlag')
plt.show()
```

```python
sns.barplot(x="Sales",y="Profit",data=iris,color='orange')
plt.show()
```

## SeaBorn Histogram / Distplot


```python
iris=pd.read_csv('../input/iris-dataset-new/iris.csv')
iris.head()
```

```python
sns.distplot(iris['Sales'])
plt.show()
```

```python
sns.distplot(iris['Price'],hist=False)
plt.show()
```

```python
sns.distplot(iris['Profit'],color="green")
plt.show()
```

```python
sns.distplot(iris['Profit'],color="green",bins=20,kde=False)
plt.show()
```

```python
sns.distplot(iris['Profit'],color="red",vertical=True)
plt.show()
```

## SeaBorn Joint Plot


```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt

iris=pd.read_csv('../input/iris-dataset-new/iris.csv')
iris.head()
```

```python
sns.jointplot(x="Sales",y="Profit",data=iris)
plt.show()
```

```python
sns.jointplot(x="Sales",y="Profit",data=iris,color="olive")
plt.show()
```

```python
sns.jointplot(x="Sales",y="Profit",data=iris,color="brown",kind="reg")
plt.show()
```

## SeaBorn Box Plot


```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt

iris=pd.read_csv('../input/iris-dataset-new/iris.csv')

sns.boxplot(x="Sales",y="Profit",data=iris)
plt.show()
```
